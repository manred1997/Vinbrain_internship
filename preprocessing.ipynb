#### Extract cxr information in dataset

```
def cxr_extract(string, pattern):
    start_idx_candidate = [match.start()+len(pattern) for match in re.finditer(pattern, string)]
    # print(start_idx_candidate)
    end_idx_candidate = []
    for i in start_idx_candidate:
        end_idx_candidate.append(string.find("\n\n", i))
    cxr_info = []
    for i in range(len(start_idx_candidate)):
        cxr_info.append(string[start_idx_candidate[i]:end_idx_candidate[i]]) 
    return {pattern: " ".join(cxr_info)}
```

```
samples = []
for s in data.TEXT:
    samples.append(cxr_extract(s, "cxr"))
```

```
tmp = []
for s in samples:
    if not s['cxr']: continue
    tmp.append(s)
samples = tmp
```

```
def reg(string):
    string = re.sub("\[\**[\w\-\(\)\s]*\**\]", "", string)
    string = re.sub("\[\**\**\]", "", string)
    string = string.replace("\n", "")
    string = re.sub("\[\**[\w\-\/]*\**\]", "", string)
    # string = re.sub("[^a-zA-Z0-9\s]")
    string = re.sub("[\.\-_][\.\-_]+", ".", string)
    string = re.sub("\(\)", "", string)
    string = re.sub("\?\?*|\~\~*", "", string)
    
    string = re.sub("[:+]", " ", string)
    string = re.sub("\d", "x", string)
    string = re.sub("[\.\,#-%>@\/]", " ", string)
    string = re.sub("\\*", " ", string)
    string = re.sub("[\(\)]", " ", string)
    string = re.sub("-", " ", string)
    string = re.sub("\*", " ", string)
    string = re.sub("\s\s+", " ", string)
    
    return string
```

```
for s in samples:
    s["cxr"] = reg(s["cxr"])
```
